```python
import MySQLdb
import random
from faker import Faker
from faker.providers import phone_number,date_time,ssn,person
```

```python
fake = Faker('uk_UA')
fake.add_provider(phone_number)
fake.add_provider(date_time)
fake.add_provider(ssn)
fake.add_provider(person)
fake.add_provider(date_time)
```

```python
#  генерація пацієнтів
people = []

for i in range(1000):
    gender_idx = random.randint(0,1)
    gender = ["Ж","Ч"][gender_idx]
    item = {
        "address": fake.address(),
        "license": fake.ssn(),
        "phone": fake.phone_number(),
        "gender": gender,
        "birthdate": fake.date_of_birth(minimum_age=16, maximum_age=90),
        "children": random.randint(0,3)
    }
    if gender_idx == 0:
        item['firstname'] = fake.first_name_female()
        item['lastname'] = fake.last_name_female()
    else:
        item['firstname'] = fake.first_name_male()
        item['lastname'] = fake.last_name_male()
    people.append(item)
```

```python
sql_inserts = []
for person in people:
    sql_inserts.append("INSERT INTO `person` "\
                       "(`name`, `surname`, `children`, `birthdate`, `address`, `phone`, `gender`)"\
                       " VALUES (\"{}\", \"{}\", {}, \"{}\", \"{}\", \"{}\", \"{}\")".format(
        person['firstname'],
        person['lastname'],
        person['children'],
        person['birthdate'].isoformat(),
        person['address'],
        person['phone'],
        person['gender']
    ))
    
with open("people_1000.sql", "w") as f:
    f.write(";\n".join(sql_inserts))
```
